```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

exp = np.array([1, 2, 3, 4, 5, 6])
sal = np.array([10000, 21000, 30000, 40000, 51000, 58000])

df = pd.DataFrame({
    "Experience": exp,
    "Salary": sal
})

print(df)

plt.figure(figsize=(10, 4))
plt.scatter(exp, sal)
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.title("Experience vs Salary")
plt.grid(True)
plt.show()

X = exp.reshape(-1, 1)
y = sal

model = LinearRegression()
model.fit(X, y)

predictions = model.predict(X)

result = pd.DataFrame({
    "Experience": exp,
    "Actual Salary": y,
    "Predicted Salary": predictions
})

result["Error"] = result["Actual Salary"] - result["Predicted Salary"]
result["Squared Error"] = result["Error"] ** 2

print(result)

mse = result["Squared Error"].mean()
mae = mean_absolute_error(y, predictions)

print("\nModel Information")
print("-" * 40)
print(f"Slope (Coefficient): {model.coef_[0]:.2f}")
print(f"Intercept: {model.intercept_:.2f}")
print(f"Equation: Salary = {model.coef_[0]:.2f} × Experience + {model.intercept_:.2f}")

print("\nEvaluation Metrics")
print("-" * 40)
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")

new_exp = np.array([[7], [8], [9], [10], [15]])
new_predictions = model.predict(new_exp)

future_predictions = pd.DataFrame({
    "Experience": new_exp.flatten(),
    "Predicted Salary": new_predictions.astype(int)
})

print("\nFuture Salary Predictions")
print("-" * 40)
print(future_predictions)
```
