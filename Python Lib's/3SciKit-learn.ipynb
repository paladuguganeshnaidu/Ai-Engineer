```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier

data = {
    "Skills": [
        "python", "java", "python", "c++", "python",
        "java", "python", "c++", "python", "java"
    ],
    "Experience": [
        3, 1, 5, 2, 7,
        0, 4, 3, 6, 1
    ],
    "Degree": [
        "btech", "bca", "mtech", "btech", "mtech",
        "bca", "btech", "btech", "mtech", "bca"
    ],
    "Selected": [
        "Yes", "No", "Yes", "No", "Yes",
        "No", "Yes", "No", "Yes", "No"
    ]
}

df = pd.DataFrame(data)

skill_encoder = LabelEncoder()
degree_encoder = LabelEncoder()
target_encoder = LabelEncoder()

df["Skills"] = skill_encoder.fit_transform(df["Skills"])
df["Degree"] = degree_encoder.fit_transform(df["Degree"])
df["Selected"] = target_encoder.fit_transform(df["Selected"])
X = df[["Skills", "Experience", "Degree"]]
y = df["Selected"]

model = DecisionTreeClassifier()

model.fit(X, y)

print("Model Trained Successfully")
```

```python
new_candidate = pd.DataFrame({
    "Skills": ["python"],
    "Experience": [0],
    "Degree": ["btech"]
})

new_candidate["Skills"] = skill_encoder.transform(new_candidate["Skills"])
new_candidate["Degree"] = degree_encoder.transform(new_candidate["Degree"])
print(new_candidate)
prediction = model.predict(new_candidate)

print(target_encoder.inverse_transform(prediction)[0])
```

```python
import requests
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

API_KEY = "1f0349cbd2cf40dc941afcb15911976c"

url = f"https://newsapi.org/v2/top-headlines?country=us&pageSize=20&apiKey={API_KEY}"

response = requests.get(url)
data = response.json()

if data.get("status") != "ok":
    print(data)
    exit()

headlines = [article["title"] for article in data["articles"] if article["title"]]

X = TfidfVectorizer(stop_words="english").fit_transform(headlines)

kmeans = KMeans(n_clusters=4, random_state=42, n_init="auto")
labels = kmeans.fit_predict(X)

df = pd.DataFrame({
    "Headline": headlines,
    "Cluster": labels
})

print(df.sort_values("Cluster"))
```
