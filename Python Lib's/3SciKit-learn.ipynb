```python
from sklearn.linear_model import LinearRegression

# Create some sample data
X = [[1], [2], [3], [4], [5]]
y = [2, 4, 6, 8, 10]

# Create the model
model = LinearRegression()

# Train the model
model.fit(X, y)

# Predict a value
prediction = model.predict([[6]])

print(prediction)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier

# ---------------- Dataset ----------------

data = {
    "Skills": [
        "Python", "Java", "Python", "C++", "Python",
        "Java", "Python", "C++", "Python", "Java"
    ],

    "Experience": [
        3, 1, 5, 2, 7,
        0, 4, 3, 6, 1
    ],

    "Degree": [
        "BTech", "BCA", "MTech", "BTech", "MTech",
        "BCA", "BTech", "BTech", "MTech", "BCA"
    ],

    "Selected": [
        "Yes", "No", "Yes", "No", "Yes",
        "No", "Yes", "No", "Yes", "No"
    ]
}

df = pd.DataFrame(data)

# ---------------- Encoders ----------------

skill_encoder = LabelEncoder()
degree_encoder = LabelEncoder()
target_encoder = LabelEncoder()

df["Skills"] = skill_encoder.fit_transform(df["Skills"])
df["Degree"] = degree_encoder.fit_transform(df["Degree"])
df["Selected"] = target_encoder.fit_transform(df["Selected"])

# ---------------- Features & Target ----------------

X = df[["Skills", "Experience", "Degree"]]
y = df["Selected"]

# ---------------- Train Model ----------------

model = DecisionTreeClassifier()

model.fit(X, y)

print("✅ Model Trained Successfully")
```

```python
new_candidate = pd.DataFrame({
    "Skills": ["Python"],
    "Experience": [4],
    "Degree": ["BTech"]
})

# Encode input
new_candidate["Skills"] = skill_encoder.transform(new_candidate["Skills"])
new_candidate["Degree"] = degree_encoder.transform(new_candidate["Degree"])

# Prediction
prediction = model.predict(new_candidate)

# Decode prediction
result = target_encoder.inverse_transform(prediction)

print("Prediction :", result[0])
```
