```python
import numpy as np
data=np.array([
    [
        [86,90],
        [98,86]
    ],
    [
        [95,83],
        [87,84]
    ]
])
print(data)
print("Dimensions: ",end="")
print(data.ndim)
print(f"Shape: {data.shape}")
print(f"Array size: {data.size}")
print(f"Data type: {data.dtype}")
arr = np.array(["AI","ML","AI","ML"])
print(arr.dtype)
```

```python
# 1D Array indexing
import numpy as np
arr=np.array([10,20,30,40,50])
print(arr)
print(arr[0])
print(arr[-4])
print(arr[2:4])
print(arr[:2])
print(arr[2:])
print(arr[2:5:2])
```

```python
data = np.array([
    [22,30000,1],
    [25,45000,3],
    [30,70000,7],
    [22,30000,1],
    [25,45000,3],
    [30,70000,7],
    [28,55000,5],
    [35,90000,10]
])
salary=data[:,1]
age=data[:,0:1]
exp=data[:,:2]
print("Average Age: ",age.mean())
print("Total mothly salary paid by a company: ",salary.sum())
print("Average Salary: ",salary.mean())
print("Total Experince: ",exp.sum())
print("Average Experince: ",exp.mean())
print("Highest salary paying to an employee: ",salary.max())
print("Lowest salary paying to an employee: ",salary.min())
print("first two employees: ",*data[:2,:])
print("Salary and Experience: \n",data[:,1:])
```
