```python
def greet_user(name):
    """Display a simple greeting."""
    print(f"Hello, {name.title()}!")
greet_user("jesse")
```

```python
#this is used to count the words as the tokens in simple way.
def count_words(text):
    words=text.split()
    num_words=len(words)
    count=0
    for word in words:
        count+=len(word)
    print(f"Number of words in text: {num_words}")
    print(f"Total number of characters in text: {count}")
txt=input("Enter a text: ")
count_words(txt)
```

```python
def reverse_text(text):
    """Reverse the given text."""
    reversed_text = text[::-1]
    print(f"Reversed text: {reversed_text}")
txt=input("Enter a text to reverse: ")
reverse_text(txt)
```

```python
def reverse_words(txt):
    words=txt.split()
    li=[]
    for word in words:
        li.append(word[::-1])
    print(li)
reverse_words("Hello this is gen-ai enginner")
```

```python
def sort_by_len(txt):
    words=txt.split()
    li=[]
    liw=[]
    for word in words:
        li.append(len(word))
        liw.append(word)
    print(li)
    print(liw.sort(key=len))
    print(liw)
sort_by_len("Prompting is the feature")
```
